# Question 1

### RUNNING SUM OF ID Array

```python
class Solution(object):
    def runningSum(self, nums):
        result = []
        a = 0
        for i in nums:
            a += i
            result.append(a)
        return result

s = Solution()
print(s.runningSum([1, 2, 3, 4]))
```

# Question 2

### Shuffle the Array

```python
nums = [2,5,1,3,4,7]
n=3
a=[]
for i in range(n):
    a.append(nums[i])
    a.append(nums[i+n])
print(a)
```

```python
class Solution(object):
    def shuffle(self, nums, n):
        a=[]
        for i in range(n):
            a.append(nums[i])
            a.append(nums[i+n])
        return a
s= Solution()
print(s.shuffle([2,5,1,3,4,7],3))
```

# Question 3

### Kids With the Greatest Number of Candies

```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        n = candies
        length =len(n)
        result =[]
        
        for i in range(length):
            result.append(n[i]+extraCandies >= max(n))
        return result
s= Solution()
print(s.kidsWithCandies([12,1,12],10))
                
```
