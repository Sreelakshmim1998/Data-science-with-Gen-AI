### Question 6 - Function

```python
def is_leap(year):
    
    if year%400 == 0 :
        return True
    elif year%100 ==0 :
        return False
        
    elif year%4 ==0:
        return True
    return False 

year = int(input())
is_leap(year)
```

```python
def is_leap(year):
    return (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)
year = int(input())
print(is_leap(year))
```
