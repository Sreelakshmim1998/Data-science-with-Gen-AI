```python
class Solution(object):
    def maxProduct(self, nums):
        max1 = 0
        max2 = 0

        for n in nums:
            if n > max1:
                max2 = max1
                max1 = n
            elif n > max2:
                max2 = n

        return (max1 - 1) * (max2 - 1)
```

```python
nums = [3, 4, 5, 2]
obj = Solution()
result = obj.maxProduct(nums)
print(result)
```

```python
class Solution(object):
    def numTeams(self, rating):
        count = 0
        n = len(rating)

        for i in range(n):
            for j in range(i + 1, n):
                for k in range(j + 1, n):
                    if rating[i] < rating[j] < rating[k] or rating[i] > rating[j] > rating[k]:
                        count += 1

        return count
```

```python
rating = [2, 5, 3, 4, 1]
obj = Solution()
result = obj.numTeams(rating)
print(result)
```

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
        return sum(1 for s, e in zip(startTime, endTime) if s <= queryTime <= e)
```

```python
startTime = [1, 2, 3]
endTime = [3, 2, 7]
queryTime = 4

obj = Solution()
result = obj.busyStudent(startTime, endTime, queryTime)
print(result)
```

```python
class Solution(object):
    def numberOfSteps(self, num):
        steps = 0
        while num:
            steps += 1
            num = num >> 1 if num % 2 == 0 else num - 1
        return steps
```

```python
num = 14
obj = Solution()
result = obj.numberOfSteps(num)
print(result)
```

```python
class Solution(object):
    def countBits(self, n):
        res = []
        for i in range(n + 1):
            res.append(bin(i).count('1'))
        return res
```

```python
n = 5
obj = Solution()
result = obj.countBits(n)
print(result)
```
