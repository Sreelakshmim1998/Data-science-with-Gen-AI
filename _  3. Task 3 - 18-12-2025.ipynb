## 2. Find Numbers with Even Number of Digits

```python
nums = [555,901,482,1771]

count =0
for i in nums:
    if len(str(i)) % 2 ==0:
        count+=1
print(count)
```

```python
class Solution(object):
    def findNumbers(self, nums):
        count = 0
        for i in nums:
            if len(str(i)) % 2 == 0:
                count+=1
        return count
a=Solution()
output = a.findNumbers([555,901,482,1771])
output
```

### 3. Number of God Pairs

```python
nums = [1, 2, 3, 1, 1, 3]

count = 0
for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] == nums[j]:
            count += 1

print(count)
```

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        count = 0
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] == nums[j]:
                    count += 1
        return count

obj = Solution()
output = obj.numIdenticalPairs([1, 2, 3, 1, 1, 3])
print(output)
```

## How Many Numbers Are Smaller Than the Current Number

```python
nums = [8,1,2,2,3]
result = []
for i in range(len(nums)):
    count =0
    for j in range(len(nums)):
        if i !=j and nums[j] < nums[i]:
            count +=1
    result.append(count) 
result       
```

```python
class Solution(object):
    def smallerNumbersThanCurrent(self, nums):
        result = []
        for i in range(len(nums)):
            count =0
            for j in range(len(nums)):
                if i !=j and nums[j] < nums[i]:
                    count +=1
            result.append(count) 
        return result       
output = Solution()
output = output.smallerNumbersThanCurrent([8,1,2,2,3])
output
```

## Subtract the Product and Sum of Digits of an Integer

```python
n=234
a=0
b=1
for i in str(n):
    a += int(i)
    b *= int(i)
    c = b- a
print(c)
```
